```python
import geopandas as gpd
from shapely.geometry import Polygon
from pyproj import Proj, transform

# Coordonnées géographiques en degrés décimaux (latitude, longitude)
coordinates = [
    (37.7749, -122.4194),  # San Francisco, USA
    (48.8566, 2.3522),     # Paris, France
    (51.5074, -0.1278),     # London, UK
    (-33.8688, 151.2093),   # Sydney, Australia
    (-22.9083, -43.1964),   # Rio de Janeiro, Brazil
    (35.6895, 139.6917),    # Tokyo, Japan
    (40.7128, -74.0060)     # New York City, USA
]
```

```python
 a= [(lon, lat) for lat, lon in coordinates]

print(a)
```

```python

# Projeter les coordonnées dans le système de coordonnées UTM 32631
in_proj = Proj(init='epsg:4326')  # Système de coordonnées géographiques (WGS 84)
out_proj = Proj(init='epsg:32631')  # UTM 32631

projected_coordinates = [transform(in_proj, out_proj, lon, lat) for lat, lon in coordinates]

# Créer un polygone avec les coordonnées projetées
polygon = Polygon(projected_coordinates)

# Créer un GeoDataFrame avec le polygone
gdf = gpd.GeoDataFrame(geometry=[polygon], crs='EPSG:32631')

# Afficher le GeoDataFrame
gdf.explore()
```
